```python
import numpy as np
import matplotlib.pyplot as plt

# Data
wtemp = np.array([330.4, 347.9, 487.3, 553.3, 643.8, 771.7, 883.5, 987.1])
wcoeff = np.array([4.35e-6, 4.47e-6, 4.61e-6, 4.72e-6,
                   4.80e-6, 4.76e-6, 4.98e-6, 4.99e-6])

zrntemp = np.array([290, 380, 480, 580, 680, 720, 800, 850, 900, 950])
zrncoeff = np.array([6.0e-6, 6.0e-6, 6.0e-6, 6.0e-6, 6.0e-6,
                     6.0e-6, 7.0e-6, 7.0e-6, 7.0e-6, 7.0e-6])

mtemp = np.array([475.85, 576.45, 674.95, 720.85, 773.15, 871.85])
mcoeff = np.array([4.37e-6, 4.78e-6, 5.04e-6, 5.47e-6, 4.35e-6, 4.52e-6])

# Volume fractions
V_Mo = 0.32
V_W  = 0.08
V_ZrN = 0.60

# Elastic moduli (GPa)
E_Mo = 320
E_W  = 407
E_ZrN = 460

# Common temperature range
T_min = max(wtemp.min(), zrntemp.min(), mtemp.min())
T_max = min(wtemp.max(), zrntemp.max(), mtemp.max())

T_common = np.linspace(T_min, T_max, 300)

# Interpolate constituent CTEs
C_W   = np.interp(T_common, wtemp, wcoeff)
C_Mo  = np.interp(T_common, mtemp, mcoeff)
C_ZrN = np.interp(T_common, zrntemp, zrncoeff)

# In-plane CTE
numerator = (
    V_Mo * E_Mo * C_Mo +
    V_W  * E_W  * C_W  +
    V_ZrN * E_ZrN * C_ZrN
)

denominator = (
    V_Mo * E_Mo +
    V_W  * E_W  +
    V_ZrN * E_ZrN
)

C_compositeip = numerator / denominator

plt.figure()
plt.plot(wtemp, wcoeff, 'o', label='W')
plt.plot(mtemp, mcoeff, '^', label='Mo')
plt.plot(zrntemp, zrncoeff, 's', label='ZrN')
plt.plot(
    T_common,
    C_compositeip,
    linewidth=2.5,
    label='MoW-ZrN'
)

plt.xlabel("Temperature (K)")
plt.ylabel("Thermal Expansion Coefficient (1/K)")
plt.title("In-Plane CTE vs Temperature")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load data
file_path = "/Users/noahmoore/Desktop/School/Thesis/TGA_NoahMoore_ZrN_1000C_10CRamp_50ml-minN2 copy.txt"

columns = [
    "Time",
    "Temperature",
    "Mass",
    "Heat Flow",
    "Diff1",
    "Diff2",
    "GasFlow",
    "Deriv Weight",
    "Mass corrected"
]

df = pd.read_csv(
    file_path,
    sep=r"\s+",
    header=None,
    names=columns,
    encoding="utf-16"
)

# Extract & convert
time = df["Time"].values * 60.0            # s
T_K = df["Temperature"].values + 273.15    # K
heat_flow = df["Heat Flow"].values * 1e-3  # mW â W
mass = df["Mass"].values * 1e-6             # mg â kg

# Heating rate
dTdt = np.abs(np.gradient(T_K, time))      # K/s

# Cp(T) 
Cp_mass = -heat_flow / (mass * dTdt)       # J/kgÂ·K

# Clean Cp
mask = np.isfinite(Cp_mass) & (Cp_mass > 0)
T_K = T_K[mask]
Cp_mass = Cp_mass[mask]

# ZrN thermal diffusivity a(T)
a_ZrN = (
    -6.89e-13 * T_K**2
    + 3.74e-9 * T_K
    + 2.14e-6
)  # m^2/s

# ZrN thermal conductivity
rho_ZrN = 7090  # kg/m^3
k_ZrN = a_ZrN * rho_ZrN * Cp_mass  # W/mÂ·K

# Literature thermal conductivities of Mo and W
k_W = (-3e-9 * T_K**3 + .0001 * T_K**2 - .277*T_K + 250.21)
k_Mo = (-3e-8 * T_K**3 + 1e-4* T_K**2 - .1243*T_K + 163.7)

# Volume fractions
f_Mo = 0.32
f_W = 0.08
f_ZrN = 0.60

# Composite thermal conductivity from rule of mixtures
k_composite = 1.0 / (
    f_Mo / k_Mo +
    f_W  / k_W +
    f_ZrN / k_ZrN
)

# Plot
plt.figure()

plt.plot(T_K, k_ZrN, label="ZrN")
plt.plot(T_K, k_Mo, label="Mo")
plt.plot(T_K, k_W, label="W")
plt.plot(T_K, k_composite, linewidth=2.5, label="MoW-ZrN")

plt.xlabel("Temperature (K)")
plt.ylabel("Thermal Conductivity (W/mÂ·K)")
plt.title("Composite Thermal Conductivity Using Measured Cp(T)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
coeffs = np.polyfit(T_K, k_composite, deg=3)
k_fit = np.polyval(coeffs, T_K)

# R^2

residuals = k_composite - k_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((k_composite - np.mean(k_composite))**2)

R2 = 1 - ss_res / ss_tot

print("Quadratic fit coefficients:", coeffs)
print("R^2 =", R2)

plt.figure()

plt.plot(T_K, k_composite, label="MoW-ZrN (data)")
plt.plot(T_K, k_fit, '--', linewidth=2,
         label=f"Cubic fit (RÂ² = {R2:.4f})")

plt.xlabel("Temperature (K)")
plt.ylabel("Thermal Conductivity (W/mÂ·K)")
plt.title("MoW-ZrN Thermal Conductivity with Cubic Fit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt

T_threshold = 650  # K

mask = T_K >= T_threshold
T_fit = T_K[mask]
k_fit_data = k_composite[mask]

def plateau_model(T, k_inf, A, T0):
    return k_inf - A * np.exp(-T / T0)


# Initial guesses
initial_guess = [
    np.max(k_fit_data),  # k_inf
    np.max(k_fit_data) - np.min(k_fit_data),  # A
    200                  # T0 (K)
]

params, _ = curve_fit(
    plateau_model,
    T_fit,
    k_fit_data,
    p0=initial_guess
)

k_inf, A, T0 = params
k_plateau_fit = plateau_model(T_fit, k_inf, A, T0)

residuals = k_fit_data - k_plateau_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((k_fit_data - np.mean(k_fit_data))**2)

R2_plateau = 1 - ss_res / ss_tot

print(f"k_inf = {k_inf:.2f} W/mÂ·K")
print(f"T0 = {T0:.1f} K")
print(f"R^2 = {R2_plateau:.4f}")
print(A)
plt.figure()

plt.plot(T_K, k_composite, label="Composite (data)")
plt.plot(
    T_fit,
    k_plateau_fit,
    '--',
    linewidth=2.5,
    label=f"Exponential Fit (RÂ² = {R2_plateau:.4f})"
)

plt.axvline(
    T_threshold,
    linestyle=':',
    linewidth=1.5,
    label="Fit threshold"
)

plt.xlabel("Temperature (K)")
plt.ylabel("Thermal Conductivity (W/mÂ·K)")
plt.title("MoW-ZrN Thermal Conductivity with Exponential Fit")
plt.legend()
plt.grid(True)
plt.show()
```
